# Population Receptive Field Estimation Technique Simulation

The aim of this computational neuroscience research project is to develope and simulate novel neuroimaging techniques for estimating PRF of human visual field using simulation.
The outline of the work can be summed up as such,

* Simulation Data
    * Creating moving bar stimulus
    * Applying size varying receptive fields
    * Retinotopic Mapping (optional)
    * Modelling BOLD Response
    * Downsampling to fMRI resolution
    
* Estimate the assumed PRF from simulated fMRI data
* Compare Assumed model with estimations
* Repeat with non-linear HRF
* Repeat with better stimulus protocol


# 1. Simulation Data
First we are going to generate the stimulus protocol.

## 1.1. Creating bar stimulus 3D array

```python
radius = 10.1
precision = 0.1
barWidth = radius / 4
angles = [-90, 45, -180, 315, 90, 225, 0, 135]
nFrames = len(angles)*3
TRs = 5 # number of TRs for each frame
TR = 3.0


X = Y = np.arange(-radius, radius , precision)
x, y = np.mgrid[-radius:radius:precision, -radius:radius:precision]
pos = np.dstack((x, y))
length = len(X)
```

```python
stim = np.zeros((nFrames, length, length))
stim_slow = np.zeros((nFrames*TRs, length, length))
f = 0
for angle in angles:
    for k in range(3):
        for i, x in enumerate(X):
            for j, y in enumerate(Y):
                if np.sin(np.deg2rad(angle))*x - np.cos(np.deg2rad(angle))*y <= (0.5-0.5*k)*radius + barWidth:
                    if np.sin(np.deg2rad(angle))*x - np.cos(np.deg2rad(angle))*y >= (0.5-0.5*k)*radius - barWidth:
                        stim[f, j, i] = 1
                        for TR in range(TRs):
                            stim_slow[f*TRs + TR, j, i] = 1
        # updates the frame
        f = f + 1
```

### Description of stimulus
Drifting bar apertures at various orientations used in Dumoulin(2007) paper; The bar width subtended 1/4th of the stimulus radius. Four bar orientations and two different motion directions for each bar were used, giving a total of 8 different bar with 3 diffrent location which gives 24 configurations within a given scan as shown bellow.

## 1.2. Receptive Field Model
For the V1 (primary visual cortex) we can model the recptive fields with a normal distribution. The probability density function for multivariate_normal is
$$
f(x) = \frac{1}{\sqrt{(2 \pi)^k \det \Sigma}} \exp\left( -\frac{1}{2} (x - \mu)^T \Sigma^{-1} (x - \mu) \right),
$$

where $\mu$ is the mean, $\Sigma$ the covariance matrix, and k is the dimension of the space where x takes values.

```python
def pRF_model(x, y, sigma, pos):
  mean = np.array([x, y])
  cov = np.diag([sigma, sigma])
  rf = multivariate_normal(mean=mean, cov=cov)
  model = rf.pdf(pos)
  return model
```

## 1.3. Applying receptive fields
First we need to assume size varying receptive field models at every point of the visual field. Then we multiply a Gaussian model with scale parameter which is a function of x and y at every point of the visual feild and then aggregating the results.

For the scale function, we use log-polar to emulate the effect of cortical zooming which are shown to simulate the foveal vision.

```python
def pRF_size(x, y):
  return 2.0*np.log(np.sqrt(np.e+x**2+y**2))

def apply_pRF(par): 
  x = par[0]
  y = par[1]
  model = pRF_model(x, y, pRF_size(x, y), pos)
  return stim_slow*model
```

```python
stim_pRF = np.zeros((nFrames*TRs, length, length))

precision_pRF = 0.5
X_pRF = Y_pRF = np.arange(-radius, radius , precision_pRF)


#Generate a list of tuples where each tuple is a combination of parameters.
#The list will contain all possible combinations of parameters.
Xs = (x for i in X_pRF)
Ys = (y for i in Y_pRF)
paramlist = list(itertools.product(Xs,Ys))

#Generate processes equal to the number of cores
pool = multiprocessing.Pool(2)

#Distribute the parameter sets evenly across the cores
res  = pool.map(apply_pRF,paramlist)
# for x in X_pRF:
#   for y in Y_pRF:
#     stim_pRF += apply_pRF(x, y)
```

## 1.4. Hemodynamic Responses
We can model the Hemodynamic response by assuming an LTI system and use the signle or double Gamma function as its impulse response. 
We can define the gamma distribution by providing its PDF: 
$$
f_X(x) = \frac{\lambda^{\alpha} x^{\alpha-1} e^{-\lambda x}}{\Gamma(\alpha)} \hspace {5pt} x > 0\\
\Gamma(\alpha) = \int_0^\infty x^{\alpha - 1} e^{-x} {\rm d}x, \hspace{20pt} \textrm{for }\alpha>0.
$$

Specifically, if $n \in \{1,2,3,...\}$, then
$$
\Gamma(n) = (n-1)!
$$


Therefore we can use a single shifted Gamma distribution with shape parameter n and rate $\lambda$ as the impulse reponse for the hemodynamic system,
$$
h(t) = \frac{\lambda^{n} (t-t_0)^{n-1} e^{-\lambda (t-t_0)}}{(n-1)!}
$$


```python
def hrf_single_gamma(t,n,lmbd,t0):
    return gamma.pdf(t,n,loc=t0,scale=lmbd)


def hrf_double_gamma(t,n1,n2,lmbd1,lmbd2,t01,t02,a):
    c = (gamma.cdf(t[t.size-1],n1,loc=t01,scale=lmbd1) 
        - a * gamma.cdf(t[t.size-1],n2,loc=t02,scale=lmbd2))
            
    return ( 1/c * (gamma.pdf(t,n1,scale=lmbd1,loc=t01) 
                   - a * gamma.pdf(t,n2,scale=lmbd2,loc=t02)) )
```

```python
# parameters for double gamma distribution function hrf:
n1 = 4
lmbd1 = 2.0
t01 = 0
n2 = 7
lmbd2 = 3
t02 = 0
a = 0.3

t = np.arange(0,nFrames*TRs*TR,TR)

hrf = hrf_double_gamma(t, n1, n2, lmbd1, lmbd2, t01, t02, a)

responses = np.zeros((nFrames*TRs, length, length))
for i in range(len(X)):
    for j in range(len(Y)):
        n = stim_pRF[:, i, j]
        responses[:, i, j] = np.convolve(hrf, n)[0:len(t)] + norm.rvs(scale=0.01, size=nFrames*TRs)
```

## 1.5. Downsampling the Hemodynamic Responses
Here we downasmple the hemodynamic responses to get the fMRI rosultion.

Let's say on average, the surface area of V1 is about 3000 $mm^3$ and the fMRI voxel sizes are 3x3x3 mm. Therefore there should be about a 1000 voxels in the fMRI data that corresponds to t30he V1 area. So let's downsample the cortical map into 30X30.

```python
nVoxels = 30
bolds = np.zeros((nFrames*TRs, nVoxels, nVoxels))
for frame in range(nFrames*TRs):
    img = responses[frame, :, :]
    img_resized = resize(img, (nVoxels, nVoxels), mode='constant', preserve_range=True)
    bolds[frame, :, :] = img_resized
```

### pRF Estimator with Total Mean Squared Error Function

```python
def estimatePRF(bolds, stim, hrf, xVoxel, yVoxel):
    def MSE(x):
      bold = bolds[:, xVoxel, yVoxel]

      model = pRF_model(x[0], x[1], x[2], pos)

      response = stim*model
      rf_response = np.sum(np.sum(response, axis=1),axis=1)
      rf_response = rf_response / np.max(rf_response) * np.max(bold)
      prediction = np.convolve(rf_response, hrf)[:bold.shape[0]]
      return np.sum((bold - prediction)**2)
    x0 = [0, 0, 1]
    bnds = ((-radius, radius), (-radius, radius), (1.0, radius))
    res = minimize(MSE, x0, bounds=bnds)
    return res
```

### Optimization

```python
res = estimatePRF(bolds, stim_slow, hrf, xVoxel=5, yVoxel=5)
```
